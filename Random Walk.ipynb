```python
# Import numpy as np
import numpy as np

# Import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
```

```python
#list for distribution of final step after each walk
all_walks_dist = []

#list for graphing all walks
all_walks_sim = []

walks = 500

#simulate random walk no.of times mentioned in range()
for i in range(walks) :
    
    # Initialize random_walk.  Start at step 0
    random_walk = [0]

    # simulate die rolls no.of times mentioned in range()
    for x in range(100) :
        # Set step: last element in random_walk
        step = random_walk[-1]

        # Roll the dice
        dice = np.random.randint(1,7)

        # Determine next step
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)

        # 10% chance of falling down all stairs due to clumsiness
        if np.random.rand() <= 0.001 :
            step = 0
            
        # append next_step to random_walk
        random_walk.append(step)
     
    #this list is to contain distrivbution of last step at end of each walk.  
    #it's not really needed, since all_walks_sim will contiain this info.   
    all_walks_dist.append(random_walk[-1])
    
    all_walks_sim.append(random_walk)

    
```

```python
#distribution of final step reached in all walks
plt.hist(all_walks_dist, bins = 10)
plt.xlabel('Final step')
plt.ylabel('No. of walks')
plt.show()
```

```python
#chances of ending at the 60th step at the very least.

#count number of walks that end at least at 60th step
count = sum(1 for x in all_walks_dist if x >= 60)

#probability
probability = count / walks
print(probability)
```

```python
#plotting simulation of all walks

# Convert all_walks to Numpy array: np_aw
np_aw_sim = np.array(all_walks_sim)

# We transpose so that correct data is in rows and columns.  See next cell for demo if we had not transposed. 
np_aw_sim_t = np.transpose(np_aw_sim)

print(np_aw_sim_t)

# Plot np_aw_t and show
plt.plot(np_aw_sim_t)

plt.xlabel('Die Roll')
plt.ylabel('Step')
        
plt.show()
```

```python
#notice how matplot lib reads the np array.  The x-values are the index positions of the sublist.
# For eg., at index 10 (x=10), the sublist (y-values) is 6, 10, 1, 13, 9. 
#These are the final steps of each walk.
```
